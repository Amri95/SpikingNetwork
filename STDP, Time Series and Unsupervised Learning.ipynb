# Introduction
Since the discovery of STDP, people have been trying very hard to understand
the computation significance of it. Indeed, there are abundant of work on the
subject, ranging from very biophysical microscopic detailed oreiented to
highly abstract and theoretical. The one by Nessler et al (2013) is an extemely
illuminating one, which provide both analytical and numerical arguments. 
Another more recent one by Dielh and Cook (2015) used somewhat more
"realistic" neuron model and has archieved better performance. However,
the common denominator of many of these (especially numerical) works is one 
ingredient: mutual inhibitions.

It has been observed that mutual inhibition between neurons from the same layer
is a common "motif" in many biological neural networks. For example, the insect
olfactory network, which primarily serves the purpose to regconize/classify
odors, has a basic multilayer feedforward architecture. Within each layers,
neurons compete with each other though either direct or indirect mutual
inhibitions. This commonly found motif is referred to a winner-take-all (WTA)
network. It has been demostrated over and over again this type of layer leads
to specialization of neurons reponds to both a particular spatial and
spatial-temporal input pattern. We would take the perspective in Nessler et al
(2013) as a starting point of our investigations. 

It is not the intention of notebook to develop the most sohpisticated
bioiphysical model of biological neural networks. Instead, we are interested
in extracting the most essential ingredient of these emergent "learning"
behavior. Quoting my colleague (who is also a pilot) Zheng Fang's saying:
"People wanted fly like birds, so they build a machine that had flapping
wings. But of course we now know, the most essential ingredient is the pair of
wings, whether they flap or not doesn't matter."



# Model Definition
##  Neuron : Integrate-and-fire model
To begin, we have a simple 2-layer feed-forward network as 
input $\mathbf{x}(t)$ and output $\mathbf{y}(t)$ respectively. 
As an input stream, $\mathbf{x}(t)$ is a binary mutlivaraite 
time-series which takes value $1$ (spike) and $0$ (rest). 
We will work in dicrete time $t$. At each moment time $t$, 
each $y_i$ receives signals from $x_j(t)$ weighted by $w_{ij}(t)$. 

$$ y_i \to y_i+ c \cdot\sum_j w_{ij}(t)x_j(t) $$
where $c$ is some constant that has to chosen properly and act as 
membrane conductance. At this moment, if any of those $y_i(t)$ is greater
than unity. This neuron $y_i$ would would be marked as to fire (spike),
and thus the name integrate-and-fire model. Futhermore, its value is reset
$$ y_i \to 0 $$


## Winner-Take-All and Mutual inhibition
Mutual inhibition inhibition between neurons $y_i$ in second layer is 
realised in the following way: 
If at any moment there are more than one neurons qualifies to spike, 
we would do a lucky draw to selcct a winner. After the winner is 
determined, a global inhibition signal $\delta_{inhib}$ would be sent to all $y_i$. 
$$ y_i \to y_i - \delta_{inhib} $$

## Spike-Timing Dependent Plasticity (STDP)
As observed in numerous experiments, the synaptic weights changes
accodring to the relative timing between pre- and post synaptic spikes.
If the pre- one preceeds the post- one, then natually this pre-synaptic
neuron has "contributed" to this particular post-spike. A "classical" 
STDP profile would lead to srengthening of such connection between 
these casually related neurons. A commonly used form for STDP is for
example:

$$\Delta w_{ij} /w_{ij} = \begin{cases}
A_{+}e^{t_i-t_j/\tau_{+}}, t_i-t_j>0\\
A_{-}e^{-(t_i-t_j)/\tau_{-}}, t_i-t_j <0
\end{cases} $$

where $t_i$ is the time of the most recent spiking event of neuron $i$.

```python
import random
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# STDP
def stdp_profile(t):
    # just a very crude one to illustrate the point
    w = 5 # stdp window
    frac = (t>=0)*1.0*np.exp(-np.clip(t/w,0.,10)) - (t<0)*.9*np.exp(np.clip(t/w,-10,0))
    return frac
ts = np.arange(-10,10)
plt.figure()
plt.plot(ts,stdp_profile(ts))
plt.title("STDP profile")
plt.xlabel("time lag")
plt.ylabel("percentage change in weight")
```

```python
# some helper function
def diff(x, y):
    if len(x) == 0 or len(y) == 0:
        return np.array([])
    else:
        xx, yy = np.meshgrid(x, y)
        return np.transpose(xx-yy)

def run_lab_stdp(inputs, weights, cond=None, inhib=0.3, train=True, leak=0.):
    weights = np.copy(weights)
    input_dim, num_time_step = inputs.shape
    num_neurons, _ = weights.shape
    if cond is None:
        cond = 100./ input_dim
    #weights_trace = np.stack(weights for i in range(num_time_step))
    weights_trace = np.empty((num_time_step, num_neurons, input_dim))
    v_array = np.zeros(num_neurons)
    last_pre_spike = np.full(input_dim, -99, dtype=np.int)
    last_pos_spike = np.full(num_neurons, -99, dtype=np.int)
    spike_recorder = np.zeros((num_neurons,num_time_step), dtype=np.int)
    inputs = cond*inputs
    for t in range(num_time_step):
        # update last spiking event
        pre_spike = np.nonzero(inputs[:,t]>0)[0]
        last_pre_spike[pre_spike] += t-last_pre_spike[pre_spike]
        # update weight_trace
        weights_trace[t,:,:] = weights
        # update membrane potential
        v_array -= leak*v_array
        v_array += weights.dot(inputs[:,t])
        # WTA, here only one lucky number is drawn to spike which would then
        # inhibit everyone
        candidates = np.nonzero(v_array>1.)[0]
        if len(candidates)>0:
            pos_spike = np.random.choice(candidates, 1)
            v_array -= inhib
            # reset spiked pos neuron
            # pdb.set_trace()
            v_array[pos_spike] = 0.0
            spike_recorder[pos_spike, t] = 1
            # update last spiking event
            last_pos_spike[pos_spike] += t-last_pos_spike[pos_spike]
            # execute STDP initiated from pos spike
            if train:
                lag = diff(last_pos_spike[pos_spike], last_pre_spike)
                frac_change = stdp_profile(lag)
                frac_change = np.squeeze(frac_change)
                #pdb.set_trace()
                weights[pos_spike, :] += weights[pos_spike, :]*frac_change
                weights[pos_spike, :] = np.clip(weights[pos_spike, :], 0., 1.)
        # execute STDP initiated from pre spike
        if train and sum(pre_spike):
            lag = diff(last_pos_spike, last_pre_spike[pre_spike])
            frac_change = stdp_profile(lag)
            weights[:, pre_spike] += weights[:, pre_spike]*frac_change
            weights[:, pre_spike] = np.clip(weights[:, pre_spike], 0., 1.)
            #pdb.set_trace()
    return spike_recorder, weights_trace, weights
```

```python
# We can visualize spiking events using a raster plot
def show_raster(input, title):
    plt.figure()
    plt.imshow(input,cmap='gray_r')
    plt.xlabel("time")
    plt.ylabel("neuron index")
    plt.title(title)
    
def show_before(inputs, weights, inhib, nrow=5, ncol=2, train=False):
    assert len(inputs) <=nrow*ncol
    fig, ax_arr = plt.subplots(nrow,ncol)
    ax_arr = ax_arr.flatten()
    for i, input in enumerate(inputs):
        spike_recorder, _, weights = run_lab_stdp(
            input, weights, inhib=inhib, train=train)
        ax_arr[i].imshow(spike_recorder,cmap='gray_r')
        ax_arr[i].text(1,1,"Input {}".format(i))
    ax_arr[-1].set_xlabel("time")
    ax_arr[-1].set_ylabel("neuron index")
    fig.suptitle("Before learning")
    plt.tight_layout(h_pad=.5)

def show_after(epochs, inputs, weights, inhib, nrow=5, ncol=2):
    assert len(inputs) <= nrow*ncol
    for epoch in range(epochs):
        for input in random.sample(inputs, len(inputs)):
            spike_recorder, _, weights = run_lab_stdp(
                input, weights, inhib=inhib, train=True)
    fig, ax_arr = plt.subplots(nrow,ncol)
    ax_arr = ax_arr.flatten()
    for i, input in enumerate(inputs):
        spike_recorder, _, weights = run_lab_stdp(
            input, weights, inhib=inhib, train=False)
        ax_arr[i].imshow(spike_recorder,cmap='gray_r')
        ax_arr[i].text(1, 1, "Input {}".format(i))
    ax_arr[-1].set_xlabel("time")
    ax_arr[-1].set_ylabel("neuron index")
    fig.suptitle("After {} epochs".format(epochs))
    plt.tight_layout(h_pad=.5)
```

# Experiment 1: stationary spatial statisitcs

```python
NUM_INPUTS = 28*28
NUM_NEURONS = 10
NUM_TIME_STEP = 40
base_rate = 0.3
inputs = []
signal = 0.5

for i in range(2):
    # background activity
    input = np.random.random((NUM_INPUTS,NUM_TIME_STEP))*(1-signal)
    # some not-trivial joint activity
    joint_neurons = np.nonzero(np.random.random(NUM_INPUTS) > 1-0.1)[0]
    joint_act = np.random.random(NUM_TIME_STEP)*signal
    for neuron in joint_neurons:
        input[neuron,:] += joint_act
    input = (input > 1-base_rate).astype(float)
    inputs.append(input)
```

```python
show_raster(inputs[0], "Input 0")
show_raster(inputs[1], "Input 1")
np.random.seed(0)
weights = np.random.rand(NUM_NEURONS,NUM_INPUTS)
np.random.seed(1)
# This "drop-out" stratepy seems to improve convergence to single output
connectivity = np.random.rand(NUM_NEURONS,NUM_INPUTS)>0.6
weights = weights*connectivity
inhib = 1.
show_before(inputs, weights, inhib, nrow=2, ncol=1, train=False)
show_after(10, inputs, weights, inhib, nrow=2, ncol=1)
```

# Experiment 2: non-stationary spatial statisitcs

```python
NUM_INPUTS = 28*28
NUM_NEURONS = 10
NUM_TIME_STEP = 40
base_rate = 0.3
inputs = []
signal = 0.5



for i in range(2):
    # background activity
    input = np.random.random((NUM_INPUTS,NUM_TIME_STEP))*(1-signal)
    # some not-trivial joint activity
    joint_neurons = np.nonzero(np.random.random(NUM_INPUTS) > 1-0.1)[0]
    joint_neurons = np.random.choice(joint_neurons, len(joint_neurons))
    joint_act = np.random.random(NUM_TIME_STEP)*signal
    for t in range(NUM_TIME_STEP):
        #print(joint_neurons)
        # slowly varying statisitcs 
        joint_neurons[t%len(joint_neurons)] = np.random.choice(NUM_INPUTS)
        joint_neurons[t+1%len(joint_neurons)] = np.random.choice(NUM_INPUTS)
        for neuron in joint_neurons:
            input[neuron,t] += joint_act[t]
    input = (input > 1-base_rate).astype(float)
    inputs.append(input)
```

```python
show_raster(inputs[0], "Input 0")
show_raster(inputs[1], "Input 1")
```

```python
np.random.seed(0)
weights = np.random.rand(NUM_NEURONS,NUM_INPUTS)
np.random.seed(1)
# This "drop-out" stratepy seems to improve convergence to single output
connectivity = np.random.rand(NUM_NEURONS,NUM_INPUTS)>0.6
weights = weights*connectivity
inhib = 1.
show_before(inputs, weights, inhib, nrow=2, ncol=1, train=False)
plt.tight_layout()
show_after(10, inputs, weights, inhib, nrow=2, ncol=1)
```
